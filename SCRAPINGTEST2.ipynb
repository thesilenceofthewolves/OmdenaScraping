```python
#to control crawlin rate 
# see here : https://www.geeksforgeeks.org/how-to-scrape-multiple-pages-of-a-website-using-python/
from time import sleep
from random import randint
import datetime
```

```python
#creted an ampty list "links" but atted an intersting link
links = []
for i in range(1, 3): # the topic has 33 pages
    url = 'https://www.edsupportforum.com/threads/pseudoephedrine-and-why-you-shouldnt-use-it.4046295/page-{}'.format(i)
    print(url) # test to see if the link have correct page and work
    links.append(url)
    sleep(randint(2,10)) # crawling rate
#print(links)https://www.edsupportforum.com/threads/what-do-you-wish-you-could-be-eating-right-now.4435953/page-2hreads/what-do-you-wish-you-could-be-eating-right-now.4435953/page-2
```

```python
import requests
from bs4 import BeautifulSoup
#create an ampty list to host
post_numbers = []
discussions = []
time_posted = []
reactions = []
titles = []
usernames = []
date_joined = []

for link in links:
    #url = links # link replace the url of the website we want to scrap. links contains the url needed
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')
    #Post number
    bs_posts = soup.find_all('a', qid="post-number")
    for post_number in bs_posts:
        post_numbers.append(post_number.text)
    #content of discussion
    bs_discussions = soup.find_all('div', class_='message-content js-messageContent')
    for disc in bs_discussions:
        discussions.append(disc.text)
    #times = soup.find_all('time' )
    bs_times = soup.find_all(qid='post-date-time')
    for time in bs_times:
        time_posted.append(time.text)
    #Reactions to the post
    bs_reactions = soup.find_all('div', qid="post-reaction-bar")
    for react in bs_reactions:
        reactions.append(react.text)
    #titles
    bs_titles = soup.find_all('h1', qid="page-header")
    for title in bs_titles:
        titles.append(title.text)
    #usernames
    bs_usernames = soup.find_all('a', class_="username")
    for name in bs_usernames:
        usernames.append(name.text)
    # date joined
    bs_djointed = soup.find_all('div', qid='message-joined-date')
    for date in bs_djointed:
        date_joined.append(date.text)

    


   
```
